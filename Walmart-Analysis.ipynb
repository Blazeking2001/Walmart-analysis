```python
import numpy as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.histplot(data = data,x = 'Weekly_Sales', kde=True)
plt.title('Distribution of Weekly Sales')
```

```python
sns.scatterplot(data = data, x = 'Temperature', y = 'Weekly_Sales')
plt.title('Weekly Sales vs Temperature')
```

```python
plt.title('Weekly Sales: Holiday vs Non-Holiday')
sns.boxplot(data = data, x = 'Holiday_Flag', y = 'Weekly_Sales')
```

```python
plt.title('CPI vs Unemployment')
sns.scatterplot(data = data, x = 'CPI', y = 'Unemployment')
```

```python
data['Date'] = pd.to_datetime(data['Date'], format='%d-%m-%Y')
data['Date']
```

```python
data.set_index('Date',  inplace=True)
```

```python
weekly_sales = data['Weekly_Sales'].resample('w').sum()
```

```python
plt.figure(figsize=(10,8))
weekly_sales.plot()
plt.title('Weekly Sales Over Time')
```
